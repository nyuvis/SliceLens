```python
import numpy as np
import pandas as pd
```

```python
X = np.random.normal(size=(1000, 20))
```

```python
y = np.random.randint(0, 2, size=(1000, 1))
```

```python
dataset = np.hstack((X, y))
```

```python
columns = list('abcdefghijklmnopqrst') + ['label']
```

```python
df = pd.DataFrame(dataset, columns=columns)
```

```python
df['label'] = df['label'].astype(int)
```
